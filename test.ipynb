```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Book1.csv') #calculating the average salary of the employees(17 answer)
average_salary = np.mean(data['Salary'])
print("average_salary=",average_salary)
```

```python
salary_filtering = data[data['Salary'] > 60000]
print(salary_filtering) #filtering the employees with salary greater than 60000(18 answer)
```

```python
plt.bar(data['Department'], data['Salary']) #plotting the bar graph of the employees and their salary(19 answer)
plt.title('Bar graph of employees and their salary')
plt.xlabel('Department')
plt.ylabel('Salary')
```
