```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('department_salary.csv') #read the data from the csv file for the answer(book2.csv)
```

```python
plt.bar(data['Name'], data['Department']) #plotting the bar graph of the employees and their salary(20 answer)
plt.title('Bar graph for department and the name of the employees')
plt.xlabel('Name')
plt.ylabel('Department')
```

```python
salary_filtering = data[data['Salary'] > 60000] 
print(salary_filtering) #filtering the employees with salary greater than 60000(21 answer)
```

```python
department_salary_trend = data.groupby('Department')['Salary'].mean()#grouping the data by department and calculating the average salary(22 answer)
plt.figure(figsize=(10,6))
plt.plot(department_salary_trend.index, department_salary_trend.values, marker='o', linestyle='-', color='blue')
plt.title('Salary Trend by Department')
plt.xlabel('Department')
plt.ylabel('Average Salary')
plt.grid(True)
plt.show()
```
